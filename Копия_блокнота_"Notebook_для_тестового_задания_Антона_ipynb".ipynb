# Запуск основного алгоритма для нахождения цепочек для GTM

```python


# ============================
# 3. Функции расчёта скоринга
# ============================

def calculate_base_scores(df, competitor_name):
    """
    Расчет базовых баллов по типу функции (KANO) и статусу реализации для конкретного конкурента с учетом
    коэффициентов конкуренции, основанных на проценте реализации функции конкурентами.
    """
    df_scores = df.merge(function_scores, left_on='kano', right_on='Function_Type', how='left')

    meta_cols = ['segment_name', 'segment_description', 'segment_id', 'segment_sam',
                 'population', 'segment_share', 'segment_avg_revenue', 'segment_awareness',
                 '3y_market_growth',
                 'solution_name', 'solution_id', 'solution_description', 'solution_importance',
                 'function_name', 'function_id', 'kano', 'function_description',
                 'Function_Type', 'Score_If_Implemented', 'Score_If_Not_Implemented']
    competitor_cols = [col for col in df_scores.columns if col not in meta_cols and not col.endswith('_x') and not col.endswith('_y')]

    # Извлекаем коэффициенты
    low_impl_coef  = competition_coefficients.loc[competition_coefficients['Competition_Level'] == 'low', 'Implemented_Coefficient'].iloc[0]
    med_impl_coef  = competition_coefficients.loc[competition_coefficients['Competition_Level'] == 'medium', 'Implemented_Coefficient'].iloc[0]
    high_impl_coef = competition_coefficients.loc[competition_coefficients['Competition_Level'] == 'high', 'Implemented_Coefficient'].iloc[0]

    low_not_impl_coef  = competition_coefficients.loc[competition_coefficients['Competition_Level'] == 'low', 'Not_Implemented_Coefficient'].iloc[0]
    med_not_impl_coef  = competition_coefficients.loc[competition_coefficients['Competition_Level'] == 'medium', 'Not_Implemented_Coefficient'].iloc[0]
    high_not_impl_coef = competition_coefficients.loc[competition_coefficients['Competition_Level'] == 'high', 'Not_Implemented_Coefficient'].iloc[0]

    def calculate_score_and_coefficients(row):
        # Подсчёт числа конкурентов, у которых функция реализована
        total_competitors = len(competitor_cols)
        implemented_count = sum([row[col] for col in competitor_cols])
        fraction = implemented_count / total_competitors if total_competitors > 0 else 0

        # Определяем уровень конкуренции
        if fraction < 0.3:
            competition_level = 'low'
        elif fraction > 0.7:
            competition_level = 'high'
        else:
            competition_level = 'medium'

        if row[competitor_name] == 1:
            # Функция реализована данным конкурентом
            if competition_level == 'low':
                comp_coef = low_impl_coef
            elif competition_level == 'high':
                comp_coef = high_impl_coef
            else:
                comp_coef = med_impl_coef
            score = row['Score_If_Implemented'] * comp_coef
        else:
            # Функция не реализована данным конкурентом
            if competition_level == 'low':
                comp_coef = low_not_impl_coef
            elif competition_level == 'high':
                comp_coef = high_not_impl_coef
            else:
                comp_coef = med_not_impl_coef
            score = row['Score_If_Not_Implemented'] * comp_coef

        return pd.Series({
            'base_score': score,
            'competition_level': competition_level,
            'competition_coefficient': comp_coef,
            'implementation_fraction': fraction
        })

    # Применяем расчёт и получаем дополнительные столбцы
    result_columns = df_scores.apply(calculate_score_and_coefficients, axis=1)
    for col in result_columns.columns:
        df_scores[col] = result_columns[col]

    return df_scores

def adjust_scores_by_market_maturity(df_scores, market_stage):
   """
   Корректировка баллов с учетом стадии зрелости рынка и типа функции.
   """
   stage_coeffs = market_maturity_coefficients[market_maturity_coefficients['Market_Stage'] == market_stage].iloc[0]

   def apply_coeff(row):
       if row['kano'] == 'attractive':
           maturity_coef = stage_coeffs['Attractive_Coefficient']
       elif row['kano'] == 'performance':
           maturity_coef = stage_coeffs['Performance_Coefficient']
       elif row['kano'] == 'must_have':
           maturity_coef = stage_coeffs['Must_Have_Coefficient']
       else:
           maturity_coef = 1.0

       return pd.Series({
           'maturity_adjusted_score': row['base_score'] * maturity_coef,
           'maturity_coefficient': maturity_coef,
           'market_stage': market_stage
       })

   # Применяем расчёт и получаем дополнительные столбцы
   result_columns = df_scores.apply(apply_coeff, axis=1)
   for col in result_columns.columns:
       df_scores[col] = result_columns[col]

   return df_scores


def calculate_final_scores(df_scores):
    """
    Расчет итоговых баллов с сохранением промежуточных коэффициентов для отладки.
    """
    # Расчет коэффициентов
    df_scores['solution_coefficient'] = df_scores['solution_importance'] / 100

    # Вычисляем итоговый балл до нормировки
    df_scores['final_score'] = df_scores['maturity_adjusted_score'] * df_scores['solution_coefficient']

    # Сохраняем положительные баллы
    df_scores['final_score_positive'] = df_scores['final_score'].clip(lower=0)

    # Группировка по сегменту и решению с сохранением промежуточных значений
    segment_solutions = df_scores.groupby(['segment_id', 'solution_name']).agg({
        'final_score_positive': 'sum',
        'final_score': 'sum',  # суммарное значение без клиппинга, для отладки
        'solution_coefficient': 'first',
        'maturity_adjusted_score': 'first',
        'base_score': 'first',
        'function_id': 'count',  # количество функций в группе
        'segment_sam': 'first',
        'segment_name': 'first',
        'population': 'first',
        'segment_share': 'first',
        'segment_avg_revenue': 'first',
        'segment_awareness': 'first',
        '3y_market_growth': 'first'
    }).reset_index()

    segment_solutions.rename(columns={'function_id': 'total_possible'}, inplace=True)

    return segment_solutions.sort_values('final_score_positive', ascending=False)


def compute_implementation_stats(df, market_stage):
    """
    Вычисление статистики реализации функций с учётом всех коэффициентов,
    включая финальные скоры с учётом важности решения и роста рынка.
    """
    meta_columns = [
        'segment_name', 'segment_description', 'segment_id', 'segment_sam',
        'population', 'segment_share', 'segment_avg_revenue', 'segment_awareness',
        '3y_market_growth',
        'solution_name', 'solution_id', 'solution_description', 'solution_importance',
        'function_name', 'function_id', 'kano', 'function_description'
    ]

    competitor_cols = [col for col in df.columns if col not in meta_columns
                      and not col.endswith('_x')
                      and not col.endswith('_y')]

    # Расчёт баллов по конкурентам
    all_competitor_scores = pd.DataFrame()

    for competitor in competitor_cols:
        try:
            competitor_df = calculate_base_scores(df, competitor)
            competitor_df = adjust_scores_by_market_maturity(competitor_df, market_stage)
            competitor_scores = calculate_final_scores(competitor_df)

            competitor_scores = competitor_scores[['segment_id', 'solution_name', 'final_score_positive']]
            competitor_scores.columns = ['segment_id', 'solution_name', competitor]

            if all_competitor_scores.empty:
                all_competitor_scores = competitor_scores
            else:
                all_competitor_scores = all_competitor_scores.merge(
                    competitor_scores,
                    on=['segment_id', 'solution_name'],
                    how='outer'
                )
        except Exception as e:
            print(f"Ошибка при обработке конкурента {competitor}: {str(e)}")

    # Считаем среднее количество финальных баллов
    score_columns = [col for col in all_competitor_scores.columns if col not in ['segment_id', 'solution_name']]
    all_competitor_scores['avg_implemented'] = all_competitor_scores[score_columns].mean(axis=1).fillna(0)

    # Расчёт максимально возможных баллов для каждой связки сегмент-решение
    def calculate_max_possible(group_df):
        try:
            # Создаём временный датафрейм с полной реализацией всех функций
            temp_df = group_df.copy()
            temp_df['temp_competitor'] = 1

            base_df = calculate_base_scores(temp_df, 'temp_competitor')
            maturity_df = adjust_scores_by_market_maturity(base_df, market_stage)
            final_df = calculate_final_scores(maturity_df)
            return final_df['final_score_positive'].sum()
        except Exception as e:
            print(f"Ошибка при расчете максимальных баллов: {str(e)}")
            return 0

    # Расчёт максимально возможных баллов
    total_possible = df.groupby(['segment_id', 'solution_name']).apply(
        lambda x: calculate_max_possible(x)
    ).reset_index(name='total_possible')

    # Объединяем результаты
    final_stats = all_competitor_scores[['segment_id', 'solution_name', 'avg_implemented']].merge(
        total_possible,
        on=['segment_id', 'solution_name']
    )

    return final_stats

def analyze_competitor(final_df, market_stage, competitor_name):
    """
    Анализ реализации конкретного конкурента с учётом баллов.
    """
    meta_columns = [
        'segment_name', 'segment_description', 'segment_id', 'segment_sam',
        'population', 'segment_share', 'segment_avg_revenue', 'segment_awareness',
        '3y_market_growth',
        'solution_name', 'solution_id', 'solution_description', 'solution_importance',
        'function_name', 'function_id', 'kano', 'function_description'
    ]
    if competitor_name not in final_df.columns:
        available = [col for col in final_df.columns if col not in meta_columns]
        raise ValueError(f"Конкурент '{competitor_name}' не найден. Доступны: {available}")

    # Расчет базовых и скорректированных баллов для конкурента
    df_scores = calculate_base_scores(final_df, competitor_name)
    df_maturity = adjust_scores_by_market_maturity(df_scores, market_stage)
    competitor_scores = calculate_final_scores(df_maturity)

    # Получаем статистику реализации баллов по всем конкурентам
    impl_stats = compute_implementation_stats(final_df, market_stage)

    # Объединяем агрегированные баллы конкурента и статистику реализации
    final_scores = competitor_scores.merge(
        impl_stats,
        on=['segment_id', 'solution_name'],
        how='left',
        suffixes=('', '_impl')
    )

    # Расчёт относительного преимущества по баллам
    final_scores['relative_advantage'] = final_scores.apply(
        lambda row: (row['final_score_positive'] - row['avg_implemented']) / row['avg_implemented']
        if row['avg_implemented'] > 0 else row['final_score_positive'] * 2,
        axis=1
    )

    # Расчёт насыщенности сегмента по баллам
    final_scores['segment_saturation'] = final_scores.apply(
    lambda row: (row['final_score_positive'] + row['avg_implemented']) / (2 * row['total_possible_impl'])
                if row['total_possible_impl'] != 0 else 0,
    axis=1
    )

    # Расчёт относительного скора
    final_scores['relative_score'] = final_scores['relative_advantage'] * (1 - final_scores['segment_saturation'])


    # Тут я провел кое-какую подмену понятий
    final_scores['normalized_score'] = np.log1p(final_scores['relative_score'])

    final_scores['normalized_score_with_importance'] = final_scores['normalized_score'] * final_scores['solution_coefficient']

    # Финальный потенциал с учётом размера сегмента
    final_scores['sam_normalized_score'] = final_scores.apply(
        lambda row: row['segment_sam'] * np.log1p(row['relative_score']),
        axis=1
    )

    final_scores['sam_normalized_score_with_importance'] = final_scores['sam_normalized_score'] * final_scores['solution_coefficient']

    # Корректировка с учетом рыночного роста
    final_scores['sam_growth_adjusted_score'] = final_scores['sam_normalized_score_with_importance'] * (1 + final_scores['3y_market_growth'] / 100)
    final_scores['competitor'] = competitor_name

    return final_scores.sort_values('sam_growth_adjusted_score', ascending=False)


def compare_competitors(final_df, market_stage, competitors=None):
    """
    Сравнение скорингов нескольких конкурентов.
    Если список конкурентов не указан, берутся все столбцы, не относящиеся к метаданным.
    """
    meta_columns = [
        'segment_name', 'segment_description', 'segment_id', 'segment_sam',
        'population', 'segment_share', 'segment_avg_revenue', 'segment_awareness',
        '3y_market_growth',
        'solution_name', 'solution_id', 'solution_description', 'solution_importance',
        'function_name', 'function_id', 'kano', 'function_description'
    ]
    if competitors is None:
        competitors = [col for col in final_df.columns if col not in meta_columns and not col.endswith('_x') and not col.endswith('_y')]

    all_results = []
    for competitor in competitors:
        try:
            res = analyze_competitor(final_df, market_stage, competitor)
            all_results.append(res)
        except Exception as e:
            print(f"Ошибка при анализе конкурента {competitor}: {str(e)}")
    if not all_results:
        raise ValueError("Нет результатов для анализа ни одного конкурента.")
    combined = pd.concat(all_results, ignore_index=True)
    return combined


def export_top_solutions(final_df, market_stage, competitors=None):
    from google.colab import files

    results = compare_competitors(final_df, market_stage, competitors)

    debug_columns = [
        'segment_name', 'solution_name', 'segment_sam',
        'population', 'segment_share', 'segment_avg_revenue', 'segment_awareness',
        '3y_market_growth',
        'base_score', 'maturity_adjusted_score', 'solution_coefficient',
        'final_score', 'final_score_positive', 'normalized_score',
        'relative_advantage', 'segment_saturation', 'relative_score', 'normalized_score_with_importance',
        'sam_normalized_score', 'sam_normalized_score_with_importance', 'sam_growth_adjusted_score', 'competitor',
        'avg_implemented', 'total_possible'
    ]

    top_solutions = results[debug_columns]
    top_solutions = top_solutions.sort_values('sam_normalized_score_with_importance', ascending=False)

    # Форматирование для экспорта
    top_solutions['sam_normalized_score'] = top_solutions['sam_normalized_score'].apply(lambda x: f"{x:,.0f}")
    top_solutions['sam_growth_adjusted_score'] = top_solutions['sam_growth_adjusted_score'].round(2)

    # Округляем новые метрики для читаемости
    for col in ['relative_advantage', 'segment_saturation', 'relative_score']:
        if col in top_solutions.columns:
            top_solutions[col] = top_solutions[col].round(3)

    filename = f'top_solutions_{market_stage}.xlsx'
    with pd.ExcelWriter(filename) as writer:
        top_solutions.to_excel(writer, sheet_name='Top Solutions', index=False)
        worksheet = writer.sheets['Top Solutions']
        for idx, col in enumerate(top_solutions.columns):
            series = top_solutions[col]
            max_len = max(series.astype(str).apply(len).max(), len(str(series.name))) + 1
            worksheet.column_dimensions[chr(65 + idx)].width = max_len

    files.download(filename)
    return top_solutions


# ============================
# 4. Пример использования
# ============================

#market_stage = 'Innovators'
#selected_competitors = ['МКБ Россия ', 'Абсолют Банк', 'ПАО "МТС-Банк"']  # или None для анализа всех конкурентов
#selected_competitors = ['Русская рыбная компания']
#selected_competitors = ['Bereke VED']


# Запуск анализа и загрузка результатов
#results = export_top_solutions(final_df, market_stage, selected_competitors)
#results
```

```python
import pandas as pd
import numpy as np
import math

def classify_sc_new(sc):
    """
    Классифицирует Solution Importance (solution_coefficient) по фиксированным диапазонам,
    деля интервал [0,1] на 5 равных частей:
      - [0.8, 1]: Очень высокая важность
      - [0.6, 0.8): Высокая важность
      - [0.4, 0.6): Умеренная важность
      - [0.2, 0.4): Низкая важность
      - [0, 0.2): Очень низкая важность
    Для каждого диапазона возвращается словарь с пояснениями для всех четырёх категорий SWOT.
    """
    if sc >= 0.8:
        return {
            "Strength": "– Потребность является критически важной для сегмента, что обеспечивает высокий спрос."
            "/n - Сильная эмоциональная привязанность к решению проблемы.",
            "Weakness": "Сильная зависимость от одной потребности; любые изменения в её восприятии могут резко снизить спрос.",
            "Opportunity": "Возможность стать лидером в удовлетворении ключевой потребности и установить долгосрочные отношения.",
            "Threat": "Усиление конкуренции или резкое изменение потребностей может привести к потере доли рынка."
        }
    elif sc >= 0.6:
        return {
            "Strength": "– Значимая потребность, гарантирующая стабильный спрос."
            "/n – Позволяет создать конкурентное преимущество через глубокое понимание сегмента.",
            "Weakness": "– Если конкуренты предложат более инновационные решения, позиция может быть утрачена.",
            "Opportunity": "Укрепление позиций за счёт качественного удовлетворения потребности и улучшения сервиса.",
            "Threat": "– Риск потери позиций при усилении конкуренции."
        }
    elif sc >= 0.4:
        return {
            "Strength": "– Потребность оказывает влияние, но не является определяющей, что позволяет использовать дополнительные факторы при выборе продукта.",
            "Weakness": "– Недостаточно сильная мотивация для покупки может затруднить выделение продукта среди конкурентов.",
            "Opportunity": "– Возможность дифференцировать предложение, сочетая удовлетворение этой потребности с другими конкурентными преимуществами.",
            "Threat": "– Низкая стимуляция спроса, особенно если конкуренты предлагают более комплексные решения."
        }
    elif sc >= 0.2:
        return {
            "Strength": "– Продукт может опираться на другие качества или функции, снижая зависимость от данной потребности.",
            "Weakness": "– Потребность не стимулирует спрос, что может снижать общий объем продаж.",
            "Opportunity": "– Возможность переориентировать маркетинговую стратегию на другие, более значимые аспекты.",
            "Threat": "– Малый интерес клиентов, что может привести к низким продажам"
        }
    else:  # sc < 0.2
        return {
            "Strength": "– Возможность полностью использовать альтернативные факторы для привлечения клиентов.",
            "Weakness": "– Потребность практически не влияет на принятие решения о покупке, снижая значимость продукта в глазах аудитории.",
            "Opportunity": "– Перенаправление усилий на другие характеристики продукта, которые действительно важны для сегмента.",
            "Threat": "– Практически отсутствие влияния на продажи, что может привести к неэффективности стратегии.."
        }

def classify_sam_new(sam):
    """
    Классифицирует SAM (segment_sam) по фиксированным диапазонам (предполагаем, что SAM нормирован от 0 до 1):
      - [0.8, 1]: Очень большой сегмент
      - [0.6, 0.8): Большой сегмент
      - [0.4, 0.6): Умеренный сегмент
      - [0.2, 0.4): Маленький сегмент
      - [0, 0.2): Очень маленький сегмент
    Возвращает словарь с пояснениями для всех категорий SWOT.
    """

    # --- Анализ SAM ---
    unique_segments = df[['segment_name', 'segment_sam']].drop_duplicates()
    q20 = unique_segments['segment_sam'].quantile(0.2)
    q40 = unique_segments['segment_sam'].quantile(0.4)
    q60 = unique_segments['segment_sam'].quantile(0.6)
    q80 = unique_segments['segment_sam'].quantile(0.8)


    if sam >= q80:
        return {
            "Strength": "– Один из сегментов лидеров. Несут потенциал продаж благодаря большому количеству потенциальных клиентов."
            "– Возможность быстрого масштабирования и массового охвата.",
            "Weakness": "– Высокая конкуренция, так как самые крупные сегменты привлекают множество игроков."
            "– Сложности в персонализации и таргетинге для каждой подгруппы.",
            "Opportunity": "– Массовые маркетинговые кампании, экономия за счет объёма, завоевание значительной доли рынка.",
            "Threat": "- Риск насыщения рынка и ценовые войны."
        }
    elif sam >= q60:
        return {
            "Strength": "– Размер сегмента выше среднего по рынку. Значительный объём потенциальных клиентов обеспечивает стабильный спрос."
            "– Возможность гибкого позиционирования при умеренной конкуренции.",
            "Weakness": "– Несмотря на значительный размер, конкуренция всё ещё может быть высокой, что влияет на маржу.",
            "Opportunity": "– Усиление бренда и расширение доли рынка при грамотном позиционировании.",
            "Threat": "– Риск оттока клиентов к конкурентам, предлагающим более привлекательные условия."
        }
    elif sam >= q40:
        return {
            "Strength": "– Средние по размеру сегменты с достаточным объёмом для стабильного развития без чрезмерной конкуренции."
            "/n– Более легкое управление и таргетирование аудитории.",
            "Weakness": "- Потенциал роста ограничен по сравнению с большими сегментами.",
            "Opportunity": "- Возможность сосредоточиться на качественном обслуживании и персонализации, формирование лояльной клиентской базы..",
            "Threat": "– Риск стагнации, если сегмент не расширяется или конкуренция усилится."
        }
    elif sam >= q20:
        return {
            "Strength": "Маленький сегмент – возможность глубокого кастомизированного подхода"
            "/n– Низкая конкуренция в узкой нише.",
            "Weakness": "- Ограниченный объём аудитории снижает общий объём продаж.",
            "Opportunity": "- Возможность установить высокую маржу за счёт эксклюзивности предложения.",
            "Threat": "Риск, что сегмент слишком узок для устойчивого роста, зависимость от изменений в специфических потребностях.."
        }
    else:
        return {
            "Strength": "- Возможность стать уникальным лидером в узкой нише. Максимальная специализация.",
            "Weakness": "- Самые маленькие сегменты с крайне ограниченным рыноком, затрудняющий масштабирование.",
            "Opportunity": "- Возможность создания эксклюзивного, премиального предложения с высокой ценой для лояльной аудитории.",
            "Threat": "Риск неокупаемости инвестиций и узкого спроса. Высокая чувствительность к изменениям рынка."
        }

def classify_ra_new(ra, sc):
    """
    Классифицирует Relative Advantage с учетом значения solution_coefficient.
    Используем фиксированные диапазоны:
      - Полное отставание (ra <= -0.5)
      - Отставание (от -0.5 до -0.2)
      - На уровне конкурентов (jn -0.2 до 0.2)
      - Преимущество (от 0.2 до 0.5)
      - Значительное преимущество (ra >= 0.5)
    Для каждого диапазона возвращаем полный словарь с четырьмя категориями SWOT.
    """
    if ra <= -0.5:
        return {
            "Strength": "",
            "Weakness": "Полное отставание – продукт существенно уступает конкурентам, что требует радикальных изменений.",
            "Opportunity": "Возможность кардинально пересмотреть стратегию и внедрить новые решения.",
            "Threat": "Риск утраты клиентов и брендового кризиса."
        }
    elif -0.5 < ra < -0.2:
        return {
            "Strength": "",
            "Weakness": "Отставание – продукт воспринимается хуже конкурентов, что снижает конкурентоспособность.",
            "Opportunity": "Возможность постепенного повышения качества для достижения уровня конкурентов.",
            "Threat": "Риск закрепления негативного имиджа, если улучшения будут медленными."
        }
    elif -0.2 <= ra <= 0.2 :
        return {
            "Strength": "Продукт соответствует среднему показателю, обеспечивая базовую конкурентоспособность.",
            "Weakness": "Отсутствие явного преимущества затрудняет завоевание дополнительной доли рынка.",
            "Opportunity": "Возможность использовать дополнительные факторы, такие как сервис или инновации, для усиления позиций.",
            "Threat": "Легкость для конкурентов улучшить свои показатели и обойти нас."
        }
    elif 0.2 < ra < 0.5:
        return {
            "Strength": "Преимущество – продукт на уровне выше среднего, что можно использовать для усиления бренда.",
            "Weakness": "Поддержание преимущества может требовать значительных инвестиций, создавая риск завышенных ожиданий.",
            "Opportunity": "Возможность закрепить лидерство, инвестируя в инновации и качество.",
            "Threat": "Конкуренты могут быстро нивелировать преимущество, снижая его эффект."
        }
    else:  # ra >= 0.5
        return {
            "Strength": "Значительное преимущество – продукт существенно превосходит средние показатели, что является мощным конкурентным драйвером.",
            "Weakness": "Поддержание такого уровня требует постоянных инвестиций и инноваций.",
            "Opportunity": "Возможность монополизировать рынок и установить новые стандарты качества.",
            "Threat": "Высокая вероятность, что конкуренты быстро адаптируются, резко снижая преимущество."
        }

def classify_sat_new(sat):
    """
    Классифицирует Segment Saturation по фиксированным диапазонам:
      - sat < 0.3: Низкая насыщенность
      - 0.3 <= sat < 0.4: Развивающийся сегмент
      - 0.4 <= sat < 0.6: Умеренно зрелый сегмент
      - 0.6 <= sat < 0.8: Зрелый сегмент
      - sat >= 0.8: Полная насыщенность
    Возвращает словарь с четырьмя категориями SWOT.
    """
    if sat < 0.3:
        return {
            "Strength": "",
            "Weakness": "",
            "Opportunity": "Низкая насыщенность – рынок ещё не заполнен, что открывает значительный потенциал для роста.",
            "Threat": "Может свидетельствовать о слабом спросе или недостатке осведомлённости."
        }
    elif sat < 0.4:
        return {
            "Strength": "",
            "Weakness": "",
            "Opportunity": "Развивающийся сегмент – часть возможностей уже освоена, что свидетельствует о формировании спроса.",
            "Threat": "Большая часть потенциала остаётся неиспользованной."
        }
    elif sat < 0.6:
        return {
            "Strength": "Умеренно зрелый сегмент – стабильный спрос и зрелость рынка.",
            "Weakness": "Частичная насыщенность может ограничивать резкий рост.",
            "Opportunity": "Возможность дальнейшего улучшения предложения за счёт оптимизации.",
            "Threat": "Риск медленного роста при усилении конкуренции."
        }
    elif sat < 0.8:
        return {
            "Strength": "Зрелый сегмент – большинство возможностей освоено, подтверждая высокий уровень конкуренции и осведомлённость аудитории.",
            "Weakness": "Ограниченное пространство для дальнейшего роста.",
            "Opportunity": "Возможность удержания лидерства за счёт постоянных обновлений.",
            "Threat": "Риск стагнации и усиления конкуренции."
        }
    else:
        return {
            "Strength": "Полная насыщенность – сегмент полностью освоен, что подтверждает зрелость рынка.",
            "Weakness": "Практически нет пространства для роста, что является слабой стороной.",
            "Opportunity": "Возможность удерживать лидерство при постоянном совершенствовании.",
            "Threat": "Риск устаревания продукта и сильное давление конкуренции."
        }

def aggregate_swot_new(row):
    """
    Объединяет вложенные словари по параметрам:
    - swot_sam_new, swot_sc_new, swot_ra_new, swot_sat_new,
    а также swot_ra_dc_combined (если он есть) в один итоговый словарь SWOT.
    Итоговый словарь имеет ключи: Strength, Weakness, Opportunity, Threat.
    Все текстовые пояснения для каждой категории объединяются через разделитель "; ".
    """
    swot = {"Strength": [], "Weakness": [], "Opportunity": [], "Threat": []}

    # Функция для объединения вложенных словарей
    def add_dict(d):
        for key, text in d.items():
            # Включаем все категории, даже Neutral, если нужно (либо пропускаем Neutral)
            if text and key != "Neutral":
                swot[key].append(text)

    add_dict(row["swot_sam_new"])
    add_dict(row["swot_sc_new"])
    add_dict(row["swot_ra_new"])
    add_dict(row["swot_sat_new"])
    # Если swot_ra_dc_combined присутствует, включаем его тоже
    if "swot_ra_dc_combined" in row and pd.notnull(row["swot_ra_dc_combined"]):
        add_dict(row["swot_ra_dc_combined"])

    # Объединяем списки в строку
    for k in swot:
        if swot[k]:
            swot[k] = "; ".join(swot[k])
        else:
            swot[k] = "None"
    return swot

def add_swot_analysis_new(df):
    """
    Новая версия функции add_swot_analysis, которая использует фиксированные диапазоны для каждого параметра:
      1. SAM (segment_sam) – через равные интервалы: [0,0.2), [0.2,0.4), [0.4,0.6), [0.6,0.8), [0.8,1]
      2. Solution Importance (solution_coefficient) – через равные интервалы: [0,0.2), [0.2,0.4), [0.4,0.6), [0.6,0.8), [0.8,1]
      3. Relative Advantage (relative_advantage) – фиксированные диапазоны: <= -1, (-1, 0), 0, (0, 1), >= 1
      4. Segment Saturation (segment_saturation) – фиксированные интервалы: <0.3, [0.3,0.4), [0.4,0.6), [0.6,0.8), >=0.8
    Результаты каждого анализа сохраняются в промежуточные столбцы:
      - swot_sam_new, swot_sc_new, swot_ra_new, swot_sat_new
    Итоговый словарь SWOT объединяет результаты всех параметров (а также swot_ra_dc_combined, если он есть) и разбивается на столбцы:
      - Strength, Weaknesses, Opportunities, Threats, swot_final_text
    """
    # --- Анализ SAM с фиксированными интервалами ---
    def classify_sam_new(sam):

        unique_segments = df[['segment_name', 'segment_sam']].drop_duplicates()
        q20 = unique_segments['segment_sam'].quantile(0.2)
        q40 = unique_segments['segment_sam'].quantile(0.4)
        q60 = unique_segments['segment_sam'].quantile(0.6)
        q80 = unique_segments['segment_sam'].quantile(0.8)

        if sam >= q80:
            return {
                "Strength": "Сегмент очень большой – огромный потенциал продаж и масштабирования.",
                "Weakness": "Сегмент очень большой – высокая конкуренция и сложности в персонализации.",
                "Opportunity": "Возможность завоевать значительную долю рынка через масштабные кампании.",
                "Threat": "Риск насыщения рынка и ценовых войн."
            }
        elif sam >= q60:
            return {
                "Strength": "Сегмент большой – стабильный спрос и гибкость позиционирования.",
                "Weakness": "Могут сохраняться проблемы с маржой из-за конкуренции.",
                "Opportunity": "Возможность укрепить бренд и расширить долю рынка.",
                "Threat": "Риск оттока клиентов к конкурентам."
            }
        elif sam >= q40:
            return {
                "Strength": "Сегмент умеренный – достаточный объём для стабильного развития.",
                "Weakness": "Ограниченный потенциал роста по сравнению с большими сегментами.",
                "Opportunity": "Возможность сфокусироваться на качественном обслуживании и персонализации.",
                "Threat": "Риск стагнации, если конкуренция усилится."
            }
        elif sam >= q20:
            return {
                "Strength": "Сегмент маленький – возможность глубокого кастомизированного подхода.",
                "Weakness": "Ограниченный объём аудитории снижает общий объём продаж.",
                "Opportunity": "Возможность установить высокую маржу за счёт эксклюзивности.",
                "Threat": "Сегмент может быть слишком узким для устойчивого роста."
            }
        else:
            return {
                "Strength": "Возможность стать уникальным лидером в узкой нише.",
                "Weakness": "Сегмент очень маленький – крайне ограниченный рынок.",
                "Opportunity": "Создание эксклюзивного, премиального предложения.",
                "Threat": "Риск неокупаемости инвестиций и узкого спроса."
            }

    df["swot_sam_new"] = df["segment_sam"].apply(classify_sam_new)

    # --- Анализ Solution Importance с фиксированными интервалами ---
    def classify_sc_new(sc):

        if sc >= 0.8:
            return {
                "Strength": "Потребность очень высока – обеспечивает высокий спрос и эмоциональную привязанность.",
                "Weakness": "Зависимость от единственной критической потребности может быть рискованной.",
                "Opportunity": "Возможность стать лидером и установить долгосрочные отношения с клиентами.",
                "Threat": "Резкое изменение восприятия может привести к потере доли рынка."
            }
        elif sc >= 0.6:
            return {
                "Strength": "Потребность высока – гарантирует стабильный спрос и конкурентное преимущество.",
                "Weakness": "Риск утраты позиции при появлении инновационных решений конкурентов.",
                "Opportunity": "Возможность укрепить позиции за счёт качественного удовлетворения потребности.",
                "Threat": "Усиление конкуренции может негативно сказаться на позициях."
            }
        elif sc >= 0.4:
            return {
                "Strength": "Потребность умеренная – оказывает влияние, но не является определяющей.",
                "Weakness": "Недостаточная мотивация для покупки затрудняет выделение продукта.",
                "Opportunity": "Возможность дифференцировать предложение в сочетании с другими преимуществами.",
                "Threat": "Низкая стимуляция спроса при наличии комплексных решений конкурентов."
            }
        elif sc >= 0.2:
            return {
                "Strength": "Низкая важность – продукт может опираться на другие качества.",
                "Weakness": "Потребность не стимулирует спрос, снижая общий объём продаж.",
                "Opportunity": "Возможность переориентировать маркетинговую стратегию на более значимые аспекты.",
                "Threat": "Малый интерес клиентов может привести к низким продажам."
            }
        else:
            return {
                "Strength": "Очень низкая важность – возможность использовать альтернативные факторы для привлечения клиентов.",
                "Weakness": "Практически не влияет на принятие решения, снижая значимость продукта.",
                "Opportunity": "Перенаправление усилий на другие, более значимые характеристики.",
                "Threat": "Практически отсутствие влияния может привести к неэффективности стратегии."
            }

    df["swot_sc_new"] = df["solution_coefficient"].apply(classify_sc_new)

    # --- Анализ Relative Advantage с фиксированными диапазонами ---
    def classify_ra_new(ra, sc):
        if ra <= -0.5:
            return {
                "Strength": "",
                "Weakness": "Полное отставание – требует кардинальных изменений.",
                "Opportunity": "Возможность полностью пересмотреть стратегию.",
                "Threat": "Риск утраты клиентов и брендового кризиса."
            }
        elif -0.5 < ra <= -0.1:
            return {
                "Strength": "",
                "Weakness": "Отставание – продукт воспринимается хуже конкурентов.",
                "Opportunity": "Возможность постепенного улучшения качества и постпенного улучшения до достижения уровня конкурентов.",
                "Threat": "Риск закрепления негативного имиджа."
            }
        elif -0.1 < ra <= 0.2:
            return {
                "Strength": "- Продукт соответствует среднему уровню – базовая конкурентоспособность.",
                "Weakness": "- Отсутствие явного преимущества усложняет завоевание дополнительной доли рынка.",
                "Opportunity": "– Возможность использовать другие факторы (сервис, инновации) для усиления конкурентных позиций.",
                "Threat": "Легко для конкурентов улучшить свои показатели и обойти нас."
            }
        elif 0.2 < ra <= 0.5:
            return {
                "Strength": "- Преимущество – продукт лучше среднего, что можно использовать для усиления бренда и повышения ценовой премии.",
                "Weakness": "– Может потребоваться значительное финансирование для поддержания такого преимущества, риск завышенных ожиданий",
                "Opportunity": "– Возможность закрепить лидерство на рынке, инвестируя в дальнейшие инновации и качество.",
                "Threat": "– Быстрое нивелирование преимущества конкурентами, что снизит эффект."
            }
        else:  # ra >= 0.5
            return {
                "Strength": "– Продукт значительно превосходит средние показатели, что является мощным конкурентным преимуществом и драйвером лидерства.",
                "Weakness": "– Поддержание такого уровня требует постоянных инвестиций и инноваций, высокие ожидания клиентов, которые трудно постоянно оправдать.",
                "Opportunity": "– Возможность монополизации рынка, установления новых стандартов качества и эффективности, высокая маржа.",
                "Threat": ""
            }

    df["swot_ra_new"] = df.apply(lambda row: classify_ra_new(row["relative_advantage"], row["solution_coefficient"]), axis=1)

    # --- Анализ Segment Saturation с фиксированными интервалами ---
    def classify_sat_new(sat):
        if sat < 0.2:
            return {
                "Strength": "",
                "Weakness": "",
                "Opportunity": "- Низкая насыщенность – неиспользованный потенциал для роста.",
                "Threat": "- Может свидетельствовать о слабом спросе."
            }
        elif sat < 0.3:
            return {
                "Strength": "",
                "Weakness": "",
                "Opportunity": "– Часть возможностей уже освоена, подтверждая интерес аудитории, сегмент начинает формироваться",
                "Threat": "– Конкуренты могут быстро заполнить незанятые ниши, если мы не будем активно развиваться."
            }
        elif sat < 0.5:
            return {
                "Strength": "– Примерно половина возможностей освоена, что свидетельствует о стабильном спросе и зрелости рынка.",
                "Weakness": "– Рынок уже частично насыщен, что ограничивает резкий рост и инновационные прорывы.",
                "Opportunity": "– Возможность дальнейшего улучшения предложения за счёт оптимизации и дополнительных инвестиций.",
                "Threat": "– Риск медленного роста и усиления конкуренции, если инновации замедлятся."
            }
        elif sat < 0.7:
            return {
                "Strength": "Зрелый сегмент – подтвержден высокий спрос.",
                "Weakness": "Ограниченное пространство для роста.",
                "Opportunity": "Возможность удержания лидерства через постоянное обновление.",
                "Threat": "Риск стагнации и усиления конкуренции."
            }
        else:
            return {
                "Strength": "Полная насыщенность – сегмент полностью освоен.",
                "Weakness": "Практически нет пространства для роста.",
                "Opportunity": "",
                "Threat": "Риск устаревания продукта и высокое давление конкуренции."
            }

    df["swot_sat_new"] = df["segment_saturation"].apply(classify_sat_new)

    # --- Сохраняем swot_ra_dc_combined как есть (если он присутствует) ---
    # Предполагаем, что если столбец 'swot_ra_dc_combined' уже есть в df, его мы оставляем.
    # Если его нет, он не добавляется.

    # --- Итоговая агрегация SWOT ---

    def aggregate_swot_new(row):
        swot = {"Strength": [], "Weakness": [], "Opportunity": [], "Threat": []}
        def add_dict(d):
            for key, text in d.items():
                # Добавляем все, кроме Neutral
                if text and key != "Neutral":
                    swot[key].append(text)
        add_dict(row["swot_sam_new"])
        add_dict(row["swot_sc_new"])
        add_dict(row["swot_ra_new"])
        add_dict(row["swot_sat_new"])
        if "swot_ra_dc_combined" in row and pd.notnull(row["swot_ra_dc_combined"]):
            add_dict(row["swot_ra_dc_combined"])
        for k in swot:
            if swot[k]:
                swot[k] = "- " + "\n- ".join(swot[k])
            else:
                swot[k] = "None"
        return swot

    df["swot_final"] = df.apply(aggregate_swot_new, axis=1)

    # Разбиваем swot_final на 4 отдельных столбца
    df["Strength"] = df["swot_final"].apply(lambda d: d.get("Strength", "None"))
    df["Weaknesses"] = df["swot_final"].apply(lambda d: d.get("Weakness", "None"))
    df["Opportunities"] = df["swot_final"].apply(lambda d: d.get("Opportunity", "None"))
    df["Threats"] = df["swot_final"].apply(lambda d: d.get("Threat", "None"))

    # Текстовое представление итогового SWOT
    def swot_to_text(swot):
        return (f"Strength: {swot.get('Strength', 'None')}. "
                f"Weaknesses: {swot.get('Weakness', 'None')}. "
                f"Opportunities: {swot.get('Opportunity', 'None')}. "
                f"Threats: {swot.get('Threat', 'None')}.")
    def swot_to_text(swot):
        return (f"Strength:\n- {swot.get('Strength', 'None')}\n"
                f"Weaknesses:\n- {swot.get('Weakness', 'None')}\n"
                f"Opportunities:\n- {swot.get('Opportunity', 'None')}\n"
                f"Threats:\n- {swot.get('Threat', 'None')}")

    df["swot_final_text"] = df["swot_final"].apply(swot_to_text)

    return df

# Пример использования:
df_analysis_with_SWOT = add_swot_analysis_new(final_analisys_final)
#df_analysis_with_SWOT[['segment_id', 'segment_name', 'solution_name', 'final_score_positive', 'final_score',
#    'solution_coefficient', 'maturity_adjusted_score', 'base_score', 'total_possible', 'segment_sam',
#    '3y_market_growth', 'avg_implemented', 'total_possible_impl', 'relative_advantage', 'segment_saturation',
#    'relative_score', 'normalized_score', 'normalized_score_with_importance', 'sam_normalized_score',
#    'sam_normalized_score_with_importance', 'sam_growth_adjusted_score', 'competitor', 'Strength',
#    'Weaknesses', 'Opportunities', 'Threats', 'swot_final_text']].drop_duplicates()
df_analysis_with_SWOT
```
